```python
with open("input.txt") as file:
    data = file.read().splitlines()
```

```python
from itertools import groupby
```

```python
elves = [list(sub) for ele, sub in groupby(data, key=bool) if ele]
```

```python
elves = [[int(cal) for cal in elve] for elve in elves]
```

```python
elvecals = [sum(elve) for elve in elves]
```

```python
elvecals.sort()
```
