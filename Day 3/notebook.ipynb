```python
def char_to_rank(input_set):
    for input_str in input_set:
        if ord(input_str) < 95:
            return ord(input_str) - 38
        else:
            return ord(input_str) - 96

def open_data_file(path):
    with open(path) as file:
        rawinput = file.read().splitlines()
    return rawinput
```

```python
data = open_data_file("input.txt")
```

```python
from collections import Counter
```

```python
def get_wrong_item(backpack):
    backpack_left = backpack[:len(backpack)//2]
    backpack_right = backpack[len(backpack)//2:]
    wrong_item = set(backpack_left).intersection(set(backpack_right))
    return wrong_item

def get_badge_item(backpack_group):
    first_elve = backpack_group[0]
    second_elve = backpack_group[1]
    third_elve = backpack_group[2]
    badge_item = set(first_elve).intersection(set(second_elve), set(third_elve))
    return badge_item
```

```python
backpack_groups = [data[i:i+3] for i in range(0, len(data), 3)]
```

```python
badge_items = [get_badge_item(backpack_group) for backpack_group in backpack_groups]
```

```python
wrong_items = [get_wrong_item(backpack) for backpack in data]
```

```python
priorities = [char_to_rank(item) for item in wrong_items]
```

```python
sum(priorities)
```

```python
priorities_badges = [char_to_rank(item) for item in badge_items]
```

```python
sum(priorities_badges)
```
